```python
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies=pd.read_csv("/kaggle/input/movyads/movies.csv")
```

```python
movies.head(3)
```

```python
movies.shape
```

```python
selected_feature=['genres','keywords','tagline','cast','director']
```

```python
for feature in selected_feature:
    movies[feature]=movies[feature].fillna('')
```

```python
combined_feature=movies['genres'] + " "+ movies['keywords']+ " " + movies['tagline'] + " " +movies['cast']+ " "+ movies['director']
print(combined_feature.head(4))
```

```python
print(combined_feature)
```

```python
vectorizer=TfidfVectorizer()
```

```python
feature_vector=vectorizer.fit_transform(combined_feature)
```

```python
print(feature_vector)
```

```python
similarity=cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
movie_name=input('Enter your movie')
```

```python
list_of_movies=movies['title'].tolist()
#print(list_of_movies)
```

```python
find_close_match=difflib.get_close_matches(movie_name,list_of_movies)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
index_of_movie=movies[movies.title==close_match]['index'].values[0]
print(index_of_movie)
```

```python
similarity_score=list(enumerate(similarity[index_of_movie]))
#print(similarity_score)
```

```python
sorted_movie=sorted(similarity_score,key=lambda x:x[1],reverse=True)
#print(sorted_movie)
```

```python
i=1
for movie in sorted_movie:
    index=movie[0]
    title_from_index=movies[movies.index==index]['title'].values[0]
    if(i<5):
        print(i,'.',title_from_index)
        i+=1
```
